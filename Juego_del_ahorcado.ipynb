## <span style="color:red"> Juego del Ahorcado versi칩n Star Wars ##

- Reglas del juego del "Ahorcado":

    - Un jugador (en este caso, el ordenador) elige una palabra secreta y dibuja un espacio para cada letra de la palabra.

    - El otro jugador intenta adivinar letras para completar la palabra.

    - Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.

    - Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

    - El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.

## <span style="color:green"> Versi칩n 1: sin interfaz gr치fica ## 


#### Funciona bien pero, de manera totalmente aleatoria, a veces se salta una letra y solo la comprueba al introducir la siguiente

```python
# importar la funci칩n random para seleccionar una palabra aleatoria de la lista de palabras
import random

# crear una lista de palabras con tem치tica Star Wars
lista_palabras = ["leia", "skywalker", "ewok", "chewbacca", "amidala", "mandalorian", "hiperespacio"]
#seleccionar una palabra aleatoria de la lista de palabras
palabra_secreta = random.choice(lista_palabras)

lista_letras = list(palabra_secreta) # Convertir la palabra secreta en una lista con cada una de sus letras
letras_unicas = set(lista_letras)   # crear un set con las letras unicas (que solo aparecen una vez) de cada palabra secreta
#print(palabra_secreta)

# dibujo de la horca como una lista de strings, cada dibujo que est치 entre las comillas triples es un string
# y al que podemos llamar con sus 칤ndices dentro de la lista

ahorcado = ['''
      +---+
      |   |
          |
          |       
          |
          |
    =========''', '''
      +---+
      |   |
      0   |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     /    |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     / \  |
          |
    =========''']




# esto inicializa el tablero de guiones para las letras lo convierte a una lista 
espacios_letras = "_"*len(palabra_secreta)
espacios_letras = list(espacios_letras)

print("Evitar que el lado oscuro te ahorque debes, joven padawan. Poderosa la fuerza en ti es.")    

# print inicial del tablero
print(" ".join(espacios_letras))


#BUCLE WHILE

# variables que nos hacen falta:
intentos = 0 # el juego admite 6 intentos hasta que el mu침eco se ahorca
#aciertos = set() # vamos a ir a침adiendo las letras acertadas a un set. As칤 si una palabra tiene varias veces una letra, solo se a침ade una vez
letras_probadas = [] # lista de las letras que se van probando para informar al jugador
jugando = True # flag para avisar al bucle while cuando se acaba el juego
letras_validas = "abcdefghijklmn침opqrstuvwxyz" # la entrada del usuario debe ser una letra


while intentos < 6 and jugando: # mientras los intentos son menos de 6 y jugando sigue siendo True:
      
      letra_usuario = input("Letra adivina ahora: ").lower() #pedirle al usuario que introduzca una letra y convertirla a minuscula
      
      if letra_usuario not in letras_validas: #para que el usuario sepa que tiene que usar letras y no otro tipo de car치cteres
            print("V치lido caracter no es. Usar letra tu debes")
            continue 


      letras_probadas.append(letra_usuario) #a침adimos la letra introducida por el usuario en la lista de letras probadas


    # Iniciar los condicionales      

      # comprobar si la letra del usuario est치 en la palabra secreta
      if letra_usuario in lista_letras:

            #para cada letra que coincida, localizar su indice y sustituir todas las instancias en la palabra secreta            
            for index, letra in enumerate(lista_letras): 
                  if letra_usuario == letra: # cuando la letra del usuario, coincide con una letra de la palabra:
                        espacios_letras[index] = letra_usuario # se sustituye el gui칩n que est치 en la posici칩n index por la letra del usuario
                
            letras_unicas.discard(letra_usuario) # con esto la eliminamos del set de letras 칰nicas, con lo que queda una letra menos que adivinar
                    
            print(" ".join(espacios_letras)) # imprimir de nuevo el tablero, y ya se ve la letra en su sitio
             
            print(f'Letras registradas por R2D2 son: {letras_probadas}') # le decimos al usuario qu칠 letras ya ha probado
                        

            # aqu칤 se verifica si todas las letras han sido adivinadas
            if len(letras_unicas) == 0: # cuando el set de letras 칰nicas est치 vac칤o es que se han adivinado todas la letras
                  jugando = False # se cambia la condici칩n que mantiene el bucle while funcionando y se sale del juego
                  print(f'La fuerza ganado ha!! {palabra_secreta.upper()}, la palabra secreta es') # esto imprime el mensaje de victoria      
      
      else: # si la letra no est치 se la palabra:
            intentos += 1 # intentos se inicializ칩 en 0, ahora queda un intento menos
            print(f'En la palabra, la letra {letra_usuario} no est치') # se imprime mensaje de que la letra no est치 en la palabra
            print(f'{6-intentos} intentos te quedan, joven padawan') # se imprime mensaje de cu치ntos intentos quedan
            # para los strings del dibujo del ahorcado, imprime el que est치 en el 칤ndice que coincide con el n칰mero de intentos
            # en el 1er intento el 칤ndice 1, que es el segundo dibujo, es decir le a침ade la cabeza, y as칤 sucesivamemte               
            print(ahorcado[intentos]) # se imprime el mu침eco del ahorcado en su estado correspondiente
            print(f'Letras registradas por R2D2 son: {letras_probadas}')
            
            

# Cuando el n칰mero de intentos llega a 6, el jugador a perdido y se le echa del juego
if intentos == 6:
      print(f'El lado oscuro ganado ha. Entrenar m치s debes, joven padawan. {palabra_secreta.upper()} la palabra secreta era') # y se le dice cual era la palabra secreta
      exit
```
