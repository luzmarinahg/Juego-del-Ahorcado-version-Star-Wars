## <span style="color:red"> Juego del Ahorcado versión Star Wars ##

- Reglas del juego del "Ahorcado":

    - Un jugador (en este caso, el ordenador) elige una palabra secreta y dibuja un espacio para cada letra de la palabra.

    - El otro jugador intenta adivinar letras para completar la palabra.

    - Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.

    - Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

    - El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.

## <span style="color:green"> Versión 1: sin interfaz gráfica ## 


#### Funciona bien pero, de manera totalmente aleatoria, a veces se salta una letra y solo la comprueba al introducir la siguiente

```python
# importar la función random para seleccionar una palabra aleatoria de la lista de palabras
import random

# crear una lista de palabras con temática Star Wars
lista_palabras = ["leia", "skywalker", "ewok", "chewbacca", "amidala", "mandalorian", "hiperespacio"]
#seleccionar una palabra aleatoria de la lista de palabras
palabra_secreta = random.choice(lista_palabras)

lista_letras = list(palabra_secreta) # Convertir la palabra secreta en una lista con cada una de sus letras
letras_unicas = set(lista_letras)   # crear un set con las letras unicas (que solo aparecen una vez) de cada palabra secreta
#print(palabra_secreta)

# dibujo de la horca como una lista de strings, cada dibujo que está entre las comillas triples es un string
# y al que podemos llamar con sus índices dentro de la lista

ahorcado = ['''
      +---+
      |   |
          |
          |       
          |
          |
    =========''', '''
      +---+
      |   |
      0   |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     /    |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     / \  |
          |
    =========''']




# esto inicializa el tablero de guiones para las letras lo convierte a una lista 
espacios_letras = "_"*len(palabra_secreta)
espacios_letras = list(espacios_letras)

print("Evitar que el lado oscuro te ahorque debes, joven padawan. Poderosa la fuerza en ti es.")    

# print inicial del tablero
print(" ".join(espacios_letras))


#BUCLE WHILE

# variables que nos hacen falta:
intentos = 0 # el juego admite 6 intentos hasta que el muñeco se ahorca
#aciertos = set() # vamos a ir añadiendo las letras acertadas a un set. Así si una palabra tiene varias veces una letra, solo se añade una vez
letras_probadas = [] # lista de las letras que se van probando para informar al jugador
jugando = True # flag para avisar al bucle while cuando se acaba el juego
letras_validas = "abcdefghijklmnñopqrstuvwxyz" # la entrada del usuario debe ser una letra


while intentos < 6 and jugando: # mientras los intentos son menos de 6 y jugando sigue siendo True:
      
      letra_usuario = input("Letra adivina ahora: ").lower() #pedirle al usuario que introduzca una letra y convertirla a minuscula
      
      if letra_usuario not in letras_validas: #para que el usuario sepa que tiene que usar letras y no otro tipo de carácteres
            print("Válido caracter no es. Usar letra tu debes")
            continue 


      letras_probadas.append(letra_usuario) #añadimos la letra introducida por el usuario en la lista de letras probadas


    # Iniciar los condicionales      

      # comprobar si la letra del usuario está en la palabra secreta
      if letra_usuario in lista_letras:

            #para cada letra que coincida, localizar su indice y sustituir todas las instancias en la palabra secreta            
            for index, letra in enumerate(lista_letras): 
                  if letra_usuario == letra: # cuando la letra del usuario, coincide con una letra de la palabra:
                        espacios_letras[index] = letra_usuario # se sustituye el guión que está en la posición index por la letra del usuario
                
            letras_unicas.discard(letra_usuario) # con esto la eliminamos del set de letras únicas, con lo que queda una letra menos que adivinar
                    
            print(" ".join(espacios_letras)) # imprimir de nuevo el tablero, y ya se ve la letra en su sitio
             
            print(f'Letras registradas por R2D2 son: {letras_probadas}') # le decimos al usuario qué letras ya ha probado
                        

            # aquí se verifica si todas las letras han sido adivinadas
            if len(letras_unicas) == 0: # cuando el set de letras únicas está vacío es que se han adivinado todas la letras
                  jugando = False # se cambia la condición que mantiene el bucle while funcionando y se sale del juego
                  print(f'La fuerza ganado ha!! {palabra_secreta.upper()}, la palabra secreta es') # esto imprime el mensaje de victoria      
      
      else: # si la letra no está se la palabra:
            intentos += 1 # intentos se inicializó en 0, ahora queda un intento menos
            print(f'En la palabra, la letra {letra_usuario} no está') # se imprime mensaje de que la letra no está en la palabra
            print(f'{6-intentos} intentos te quedan, joven padawan') # se imprime mensaje de cuántos intentos quedan
            # para los strings del dibujo del ahorcado, imprime el que está en el índice que coincide con el número de intentos
            # en el 1er intento el índice 1, que es el segundo dibujo, es decir le añade la cabeza, y así sucesivamemte               
            print(ahorcado[intentos]) # se imprime el muñeco del ahorcado en su estado correspondiente
            print(f'Letras registradas por R2D2 son: {letras_probadas}')
            
            

# Cuando el número de intentos llega a 6, el jugador a perdido y se le echa del juego
if intentos == 6:
      print(f'El lado oscuro ganado ha. Entrenar más debes, joven padawan. {palabra_secreta.upper()} la palabra secreta era') # y se le dice cual era la palabra secreta
      exit
```

```python
# Versión 2 con interfaz gráfica

import random # importar la función random para poder seleccionar una palabra aleatoria de la lista de palabras

import tkinter as tk # importar el módulo tkinter para poder hacer la interfaz
from tkinter import ttk # importar el módulo ttk para poder construir y colocar los widgets
from tkinter import messagebox # manda mensaje de error si se introduce un typeerror

# función para seleccionar una palabra secreta aleatoria con temática Star Wars
def seleccionar_palabra_aleatoria():
      lista_palabras = ["leia", "skywalker", "ewok", "chewbacca", "amidala", "mandalorian", "hiperespacio"]
      #seleccionar una palabra aleatoria de la lista de palabras
      return random.choice(lista_palabras)
      



# función con el dibujo de la horca como una lista de strings, así los podemos llamar con sus índices dentro de la lista
def mostrar_ahorcado():
      return ['''
            +-----+
            |     |
            |   
            |          
            |
            |
            =========''', '''
            +-----+
            |     |
            |     0   
            |
            |
            |
            =========''', '''
            +-----+
            |     |
            |     O   
            |     |
            |
            |
            =========''', '''
            +-----+
            |     |
            |     O   
            |    /|
            |
            |
            =========''', '''
            +-----+
            |     |
            |     O   
            |    /|\  
            |
            |
            =========''', '''
            +-----+
            |     |
            |     O   
            |    /|\  
            |    / 
            |
            =========''', '''
            +-----+
            |     |
            |     O   
            |    /|\  
            |    / \            
            =========''']

      


# función para inicializar el tablero de guiones para las letras y convertirlo en una lista
def iniciar_tablero(palabra_secreta):
      espacios_letras = "_"*len(palabra_secreta)
      return list(espacios_letras)


#FUNCIÓN PRINCIPAL, JUGAR Y COMPROBAR LAS LETRAS
def jugar():
      # estas variables tienen que ser globales para estar disponibles para todas las funciones
      global intentos, jugando, palabra_secreta, letras_unicas, letras_probadas, espacios_letras, ahorcado
      alfabeto = 'abcdefghijklmnopqrstuvwxyz'
                  
      # Se recoge la letra del usuario, en lugar de un input usamos la sintáxis del módulo tkinter
      letra_usuario = entry1.get().lower() #aquí la letra_usuario es la que se introdujo en la interfaz
      entry1.delete(0, tk.END) #esta línea elimina la letra introducida para poder introducir otra
            
      if str(letra_usuario) not in alfabeto or len(letra_usuario) != 1: #si se introduce algo que no esté en alfabeto o más de una letra
            # se muestra un mensaje de error
            messagebox.showwarning("Input Error", "Usar una sola letra del alfabeto, tú debes.")
            error_letra_label.config(text="Usar una sola letra del alfabeto, tú debes.")
            return

      if letra_usuario in letras_probadas: # si la letra ya se ha probado
            # se muestra un mensaje de error
            messagebox.showwarning("Input Error", "Letra probada ya, joven padawan.")
            error_letra_label.config(text="Letra probada ya, joven padawan.")
            return
            
      letras_probadas.append(letra_usuario) #añadimos la letra introducida por el usuario en la lista de letras probadas


      # Iniciar los condicionales                 
      if letra_usuario in lista_letras: # comprobar si la letra del usuario está en la palabra secreta

            #para cada letra que coincida, localizar su indice y sustituir todas las instancias en la palabra secreta            
            for index, letra in enumerate(lista_letras): 
                  if letra_usuario == letra: # cuando la letra del usuario, coincide con una letra de la palabra:
                        espacios_letras[index] = letra_usuario # se sustituye el guión que está en la posición index por la letra del usuario
                  
            letras_unicas.discard(letra_usuario) # eliminamos esa letra del set de letras únicas, con lo que queda una letra menos que adivinar                    
              

            # aquí se verifica si todas las letras han sido adivinadas
            if len(letras_unicas) == 0: # cuando el set de letras únicas está vacío es que se han adivinado todas la letras
                  jugando = False # se cambia la condición que mantiene el bucle funcionando para salir del juego
                  # llegado a este punto, el jugador ha ganado la partida, y se imprime un menaje de victoria
                  result_label.config(text=f'Ganado la fuerza ha 🧌⭐️🧌⭐️!! {palabra_secreta.upper()}, la palabra secreta es 😀') # convertir la palabra secreta a mayúscula para que se vea mejor
            
      else: # si la letra no está se la palabra:
            intentos += 1 # intentos se inicializó en 0, ahora queda un intento menos
            ahorcado_label.config(text=ahorcado[intentos])
            
                  
      # Cuando el número de intentos llega a 6, el jugador ha perdido y se le echa del juego
      if intentos == 6:
            jugando = False
            result_label.config(text=f'El lado oscuro ganado ha. Entrenar más debes, joven padawan 🥋🥋🥋. {palabra_secreta.upper()} la palabra secreta era') # y se le dice cual era la palabra secreta
            return
      
      tablero_label.config(text=" ".join(espacios_letras))
      letras_probadas_label.config(text=f'Letras registradas por R2D2 son: {letras_probadas}')


# variables que necesitamos para inicializar el juego:

palabra_secreta = seleccionar_palabra_aleatoria() #aquí llamamos a la función para seleccionar la palabra secreta
lista_letras = list(palabra_secreta) # Convertir la palabra secreta en una lista con cada una de sus letras
letras_unicas = set(lista_letras)   # crear un set con las letras unicas (que solo aparecen una vez) de cada palabra secreta
ahorcado = mostrar_ahorcado() # llamamos a la funcion para poder mostrar el estado del ahorcado

espacios_letras = iniciar_tablero(palabra_secreta) #iniciamos el tablero

intentos = 0 # el juego admite 6 intentos hasta que el muñeco se ahorca
letras_probadas = [] # lista de las letras que se van probando para informar al jugador
jugando = True # flag para avisar al bucle cuando se acaba el juego


# sintaxis de tkk
# Crear la ventana principal
root = tk.Tk()
root.title("⭐️⭐️⭐️ Ahorcado Star Wars ⭐️⭐️⭐️")
root.geometry("700x600")


# Crear y colocar los widgets en la ventana

tablero_label = ttk.Label(root, text=" ".join(espacios_letras), font=("Helvetica", 24)) 
tablero_label.grid(column=0, row=1, columnspan=2, pady=10)

error_letra_label = ttk.Label(root, text=' ', font=("Helvetica", 14)) #etiqueta para los mensajes de error en la ventana de error
error_letra_label.grid(column=0, row=4, columnspan=2, pady=20)

ahorcado_label = ttk.Label(root, text=ahorcado[0], font=("Helvetica", 24)) # etiqueta para colocar el dibujo del ahorcado
ahorcado_label.grid(column=0, row=3, columnspan=2, pady=10)

letras_probadas_label = ttk.Label(root, text=f'Letras registradas por R2D2 son: {letras_probadas}', font=("Helvetica", 20)) #etiqueta para mostrar las letras que ya se han utilizado
letras_probadas_label.grid(column=0, row=4, columnspan=2, pady=20)

ttk.Label(root, text="🧌⭐️: Letra introduce, joven padawan:", font=("Helvetica",24)).grid(column=0, row=0, padx=10, pady=10) # recuadro para introducir la letra
entry1 = ttk.Entry(root) #entry1 es la letra que introduce el usuario, que se manda al programa principal
entry1.grid(column=1, row=0, padx=10, pady=10)

add_button = ttk.Button(root, text="Comprueba la letra", command=jugar) # Boton para comprobar la letra, "jugar" es nuestra función principal
add_button.grid(column=0, row=2, columnspan=2, padx=10, pady=2)

result_label = ttk.Label(root, text="Ganador: ", font=("Helvetica", 20)) # etiqueta que imprime quien ha ganado al final del juego
result_label.grid(column=0, row=5, columnspan=2, padx=10, pady=10)

# Comando para ejecutar el programa completo
root.mainloop()
```
